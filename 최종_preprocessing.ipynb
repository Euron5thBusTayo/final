1. 데이터 불러오기, 합치기

```python
 import pandas as pd
 import numpy as np
 import warnings; warnings.filterwarnings('ignore')
```

```python
#월별 버스 데이터를 train 하나로 합치기
path = '/content/drive/MyDrive/Bus_Tayo/bus'
train_files = ['2022_09.csv', '2022_10.csv', '2022_11.csv', '2022_12.csv', '2023_01.csv', '2023_02.csv',
        '2023_03.csv', '2023_04.csv', '2023_05.csv', '2023_06.csv', '2023_07.csv', '2023_08.csv']
train_df = [pd.read_csv(path+'/'+f) for f in train_files]
train = pd.concat(train_df)
```

```python
train.head()
```

```python
train.info()
```

2. 불필요한 피처 삭제

등록일자, 교통수단타입코드, 버스정류장ARS번호, 노선명 칼럼을 삭제한다.

```python
train = train.drop(['등록일자', '교통수단타입코드', '버스정류장ARS번호', '노선명'], axis=1)
train.head()
```

3. 심야버스 삭제

출퇴근 시간대를 분석할 예정이므로 심야버스 row를 삭제한다.

```python
train = train[train['교통수단타입명'] != '서울심야버스']
count = train[train['교통수단타입명'] == '서울심야버스'].shape[0]
print(f'서울심야버스인 행의 개수: {count}')
```

4. 출퇴근 시간대를 제외한 시간대 column 삭제, 2시간 단위로 시간대 재구성

출퇴근 시간이 아닌 20시~5시 데이터를 삭제한다.

6-8, 8-10, 10-12, 12-14, 14-16, 16-18, 18-20, 20-22로 2시간 단위로 시간대를 묶는다.

X~Y ride(2시간 단위 승차 총 승객 수), X~Y takeoff(2시간 단위 하차 총 승객 수)로 칼럼명을 변경한다.

```python
train['6~8 ride'] = train['6시승차총승객수'] + train['7시승차총승객수']
train['8~10 ride'] = train['8시승차총승객수'] + train['9시승차총승객수']
train['10~12 ride'] = train['10시승차총승객수'] + train['11시승차총승객수']
train['12~14 ride'] = train['12시승차총승객수'] + train['13시승차총승객수']
train['14~16 ride'] = train['14시승차총승객수'] + train['15시승차총승객수']
train['16~18 ride'] = train['16시승차총승객수'] + train['17시승차총승객수']
train['18~20 ride'] = train['18시승차총승객수'] + train['19시승차총승객수']

time_drop_columns = [f'{hour}시승차총승객수' for hour in range(1,24)]
train.drop(time_drop_columns, axis=1, inplace=True)
train.drop('00시승차총승객수', axis=1, inplace=True)

train['6~8 takeoff'] = train['6시하차총승객수'] + train['7시하차총승객수']
train['8~10 takeoff'] = train['8시하차총승객수'] + train['9시하차총승객수']
train['10~12 takeoff'] = train['10시하차총승객수'] + train['11시하차총승객수']
train['12~14 takeoff'] = train['12시하차총승객수'] + train['13시하차총승객수']
train['14~16 takeoff'] = train['14시하차총승객수'] + train['15시하차총승객수']
train['16~18 takeoff'] = train['16시하차총승객수'] + train['17시하차총승객수']
train['18~20 takeoff'] = train['18시하차총승객수'] + train['19시하차총승객수']

time_drop_columns = [f'{hour}시하차총승객수' for hour in range(1,24)]
train.drop(time_drop_columns, axis=1, inplace=True)
train.drop('00시하차총승객수', axis=1, inplace=True)
```

```python
train.head()
```

5. 역명 칼럼 변환 및 칼럼 이름 재구성

역명 칼럼을 괄호 안 5자리 숫자(00000)로 대체한다. 또한, 이름을 '역번호'로 수정한다.

사용년월, 노선번호, 역명, 표준버스정류장ID, 교통수단타입명, 역번호를 영문으로 변경한다.

```python
train['역번호'] = train['역명'].str.extract(r'\((\d+)\)')
train.drop('역명', axis=1, inplace=True)
train.head()
```

```python
columns={'사용년월':'month', '노선번호': 'bus_ID', '표준버스정류장ID': 'bus_station_ID',
        '교통수단타입명':'type', '역번호': 'bus_station_ID', '표준버스정류장ID': 'bus_station_SID'}

train.rename(columns=columns, inplace=True)

train.head()
```

6. Object type 피처 처리

bus_ID는 110A, 종로09와 같이 숫자와 문자열이 함께 섞여 있다.

따라서, 숫자형으로 변환하기 쉽도록 bus_ID 대신 노선 정보 데이터 속 route_ID로 대체한다.

노선 정보 데이터 링크 : https://data.seoul.go.kr/dataList/OA-15262/S/1/datasetView.do

```python
#노선 정보 데이터 불러오기
path = '/content/drive/MyDrive/Bus_Tayo/data'
route_info = pd.read_csv(path+'/route_info.csv', encoding='utf-8')
route_info.rename(columns={'노선명': 'bus_ID', 'ROUTEID': 'route_ID'}, inplace=True)
route_info
```

```python
#bus_ID를 기준으로 하여 train과 route_info 합치기
train = pd.merge(train, route_info, on='bus_ID', how='left')

train
```

```python
#route_ID 값이 없는 bus_ID가 존재하는지 확인
null_routes_df = train[train['route_ID'].isnull()][['bus_ID']].drop_duplicates()
print("route_ID 값이 null인 bus_ID:")
print(null_routes_df)
```

버스 노선 고유 아이디(route_ID)와 매핑되지 못한 데이터가 존재한다.

수동으로 route_ID를 찾을 수 있는 경우 직접 찾아 넣었다.

매핑되지 못하며 수동으로 route_ID를 찾을 수 없는 버스는 임의로 route_ID를 부여해 처리하였다.
111110000~111110011 사이 숫자로 부여하였다.

```python
train[train['route_ID'].isnull()]['bus_ID'].count()
```

```python
#route_ID가 없는 bus_ID 처리
#검색을 통해 route_ID를 찾을 수 있는 경우 직접 찾아서 넣음
#그렇지 못한 경우 111110000~111110011 사이 숫자를 임의로 부여하여 처리함
null_routes = {'1': 100100001, '9703': 100100398, '153': 111110000, '9403': 100100390, '성동07': 103900005,
               '성동06': 103900007, '2235': 111110001, '2234': 111110002, '8112': 111110003, '1156': 100100166,
               '1157': 100100166, '노원04': 111110004, '성북10-2': 107900017, '성북10-1': 107900014,
               '은평01': 111900002, '8761': 111110005, '마포14': 113900006, '6649': 114000001,
               '6513': 111110007, '6611': 100100295, '6635': 100100311, '6654': 100100322,
                '금천01-1': 117900008, '8001': 106000004, '17': 111110008, '411': 111110009, '6511': 111110010,
               'Jan.01': 111110011}
null_routes = pd.DataFrame.from_dict(null_routes, orient='index', columns=['route_ID']).reset_index()
null_routes.columns = ['bus_ID', 'route_ID']

null_routes
```

```python
train = pd.merge(train, null_routes, on='bus_ID', how='left')
train
```

```python
#필요하지 않은 route_ID_x와 route_ID_y 칼럼 삭제
train['route_ID'] = train['route_ID_x'].fillna(train['route_ID_y'])
train.drop(['route_ID_x', 'route_ID_y'], axis=1, inplace=True)
train.insert(4, 'route_ID', train.pop('route_ID'))
```

```python
train.head()
```

```python
null_routes_df = train[train['route_ID'].isnull()][['bus_ID']]
print("route_ID 값이 null인 bus_ID:")
print(null_routes_df)
train.drop(['bus_ID'], axis=1, inplace=True)
```

route_ID와 bus_ID 전부 매핑되었다.

이어서 bus_station_ID와 bus_station_SID를 연결하여 하나의 고유키를 생성한다.
예) bus_station_ID가 00077, bus_station_SID가 100000001인 경우 bus_station_PK는 00077100000001이 된다.

```python
#고유키 생성 및 기존 칼럼 삭제
train['bus_station_PK'] = train['bus_station_ID'].astype(str) + train['bus_station_SID'].astype(str)
train.drop(['bus_station_ID'], axis=1, inplace=True)
train.drop(['bus_station_SID'], axis=1, inplace=True)
#데이터 자료형을 정수형으로 바꿈
train['bus_station_PK'] = train['bus_station_PK'].astype(int)
train.head()
```

```python
#칼럼별 자료형 확인
train.info()
```

버스 타입 카테고리를 다음과 같은 숫자로 바꾼다.

type

서울간선버스(0), 서울지선버스(1), 서울마을버스(2), 서울순환버스(3), 서울광역버스(4)

```python
type_category = {
    '서울간선버스': 0,
    '서울지선버스': 1,
    '서울마을버스': 2,
    '서울순환버스': 3,
    '서울광역버스': 4
}
train['type'].replace(type_category, inplace=True)

train.head()
```
